```python
# Import pandas and alias
import pandas as pd

# Read CSV as a DataFrame
office_df = pd.read_csv('datasets/office_episodes.csv')

# Print the top data in the DataFrame
office_df.head()
```

```python
# Define an empty list for color scheme
colors = []

# Iterate over rows of scaled_ratings from office_df with color scheme condition
for lab, row in office_df.iterrows():
    if row['scaled_ratings'] < 0.25:
        colors.append('red')
    elif 0.25 <= row['scaled_ratings'] < 0.50:
        colors.append('orange')
    elif 0.50 <= row['scaled_ratings'] < 0.75:
        colors.append('lightgreen')
    else:
        colors.append('darkgreen')
        
# Inspect the first 10 values in the list        
colors[:10]
```

```python
# Define sizing system variable list
sizes = []

# Iterate over row in office_df which has_guests marked size 250, else 25
for lab, row in office_df.iterrows():
    if row['has_guests'] == True:
        sizes.append(250)
    else:
        sizes.append(25)
        
# Inspect the first 10 values in the list        
sizes[:10]
```

```python
# Import matplotlib.pyplot with alias
import matplotlib.pyplot as plt

# Set the figure style and initalize a new figure
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(11, 7))

# Create a scatter plot of episode number vs viewership
plt.scatter(office_df['episode_number'], office_df['viewership_mil'], c=colors, marker='*', s=sizes)

# Create a title and axis labels
plt.title('Popularity, Quality, and Guest Appearances on the Office', size = 16)
plt.xlabel('Episode Number', size=12)
plt.ylabel('Viewership (Millions)', size=12)

# Show the plot
plt.show()
```
